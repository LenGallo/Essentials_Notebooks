## Problem 2d

```python
def Scalc_w_cutoff(vgps,c):
    N=len(vgps[1])
    colats=90.-vgps[1]
    Good=[]
    for i in range(N):
        if colats[i]<c:Good.append(colats[i])
    sel_colats=np.array(Good)
    Sp2=np.sum(sel_colats**2)/(N-1.)
    return '%7.1f'%(np.sqrt(Sp2))
```

```python
for c in range(90,20,-5):
    print c,Scalc_w_cutoff(VGPs,c)
```

## Problem 2e

Ok.  now we repeat this with the data in Chapter_14/hawaii\_results.txt

```python
hw_site_means=pd.read_csv('Chapter_14/hawaii_results.txt','\t',header=1)
hw_site_means=hw_site_means.ix[hw_site_means.data_type=='i']
hw_site_means=hw_site_means.dropna(subset = ['average_dec'])
hw_site_means=hw_site_means.dropna(subset = ['average_inc'])
hw_site_means['vgp_ant_lat']=-hw_site_means['vgp_lat']
hw_site_means['vgp_ant_lon']=hw_site_means['vgp_lon']-180.
hw_Norm_results=hw_site_means.ix[hw_site_means.vgp_lat>=0]
hw_Rev_results=hw_site_means.ix[hw_site_means.vgp_lat<0]
hw_VGPs=np.concatenate((hw_Norm_results[['vgp_lon','vgp_lat']].values,\
                     hw_Rev_results[['vgp_ant_lon','vgp_ant_lat']].values)).transpose()
print Scalc(hw_VGPs)
```

Wow. so there is a BIG latitudinal dependence.  Now let's calculate Model G for the latitudes at Hawaii and Antarctica.

```python
def ModelG(lat):
    a,b=0.26,11.9
    S2=(a*lat)**2+b**2
    return np.sqrt(S2)
```

```python
antarctica_lat=np.average(site_means['average_lat'])
hawaii_lat=np.average(hw_site_means['average_lat'])
print antarctica_lat,ModelG(antarctica_lat)
print hawaii_lat,ModelG(hawaii_lat)
```

So the values we get are much larger than Model G.  But what about using a cutoff of, say 45$^{\circ}$?

```python
print Scalc_w_cutoff(VGPs,45)
print Scalc_w_cutoff(hw_VGPs,45)
```

So, Model G works well for data with a 45$^{\circ}$ cutoff.  And now just to show off:

```python
lats=range(100)
Ss=[]
for lat in lats:
    Ss.append(ModelG(lat))
plt.plot(lats,Ss)
```
